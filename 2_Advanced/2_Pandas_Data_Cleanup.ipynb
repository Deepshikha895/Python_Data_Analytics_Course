```python
#importing required packages

import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading dataset
dataset=load_dataset('lukebarousse/data_jobs')
#forming a dataframe
df = dataset['train'].to_pandas()


#data cleanup converting date column to date timetype from string type.
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python

median_salary_year = df['salary_year_avg'].median()
median_salary_hour = df['salary_hour_avg'].median()
df_filled = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(median_salary_year)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(median_salary_hour)

```

```python
df_filled.loc[0:10,'salary_year_avg':'salary_hour_avg']
```

```python
df_unique = df_filled
df_unique = df_unique.drop_duplicates()

print("Length of orignal dataset: ", len(df_filled))
print("Length of unique dataset: ", len(df_unique))
print("No of dropped rows: ", len(df_filled)-len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset = ['job_title', 'company_name'])
```
