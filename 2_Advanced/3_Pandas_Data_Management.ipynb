```python
#importing required packages

import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading dataset
dataset=load_dataset('lukebarousse/data_jobs')
#forming a dataframe
df_orignal = dataset['train'].to_pandas()


#data cleanup converting date column to date timetype from string type.
df_orignal['job_posted_date'] = pd.to_datetime(df_orignal['job_posted_date'])
```

df_altered = df_orignal.copy()

```python
df_altered = df_orignal.copy()
print("Id of orignal data frame:    ", id(df_orignal))
print("ID of copied data frame:     ", id(df_altered))
print("Are both the IDS same?       ", id(df_orignal)==id(df_altered))
df_altered.head()
```

```python

df_altered = df_altered.drop_duplicates()
df_altered = df_altered.drop_duplicates(subset = ['job_title', 'company_name'])
median_salary_year = df_altered['salary_year_avg'].median()
median_salary_hour = df_altered['salary_hour_avg'].median()
df_altered ['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary_year)
df_altered['salary_hour_avg'] = df_altered['salary_hour_avg'].fillna(median_salary_hour) 
print("Length of new data frame:     ", len(df_altered))
print("Length of orignal data frame: ", len(df_orignal))
df_altered.head()
```
