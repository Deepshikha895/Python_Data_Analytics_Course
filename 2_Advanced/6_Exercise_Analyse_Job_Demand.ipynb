```python
#importing required packages

import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading dataset
dataset=load_dataset('lukebarousse/data_jobs')
#forming a dataframe
df = dataset['train'].to_pandas()


#data cleanup converting date column to date timetype from string type.
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_USA = df[df['job_country']=='United States'].copy()
df_USA
```

```python
df_USA['job_posted_month'] = df_USA['job_posted_date'].dt.strftime('%B')
df_USA
```

```python
df_us_pivot=df_USA.pivot_table(index='job_posted_month', columns= 'job_title_short', aggfunc='size')
df_us_pivot
```

```python
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B')
df_us_pivot['month_no'] = df_us_pivot['month_no'].dt.month
df_us_pivot.sort_values('month_no',inplace=True)
df_us_pivot.set_index('job_posted_month', inplace=True)
df_us_pivot.drop(columns='month_no', inplace=True)


```

```python
df_us_pivot.plot(kind='line')
```

```python
top_3=df_USA['job_title_short'].value_counts().head(3)
top_3=top_3.index.to_list()
top_3
```

```python
df_us_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postingsfor the top data jobs in the US')
plt.xlabel('Year 2023')
plt.ylabel('Job count')
plt.legend()
plt.show()
```
