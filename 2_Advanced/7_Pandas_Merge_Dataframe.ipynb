```python
#importing required packages

import pandas as pd
import numpy as np
from datasets import load_dataset
import matplotlib.pyplot as plt

#loading dataset
dataset=load_dataset('lukebarousse/data_jobs')
#forming a dataframe
df = dataset['train'].to_pandas()


#data cleanup converting date column to date timetype from string type.
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_USA = df[df['job_country']=='United States'].copy()
df_USA
```

```python
df_USA['job_posted_month'] = df_USA['job_posted_date'].dt.strftime('%B')
```

```python
df_us_pivot=df_USA.pivot_table(index='job_posted_month', columns= 'job_title_short', aggfunc='size')
df_us_pivot
```

```python
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'], format='%B')
df_us_pivot['month_no'] = df_us_pivot['month_no'].dt.month
df_us_pivot.sort_values('month_no',inplace=True)
df_us_pivot.set_index('job_posted_month', inplace=True)
df_us_pivot.drop(columns='month_no', inplace=True)
```

```python
df_us_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col = 'job_posted_month')
df_us_software_pivot
```

```python
df_us_merged = df_us_pivot.merge(df_us_software_pivot, on="job_posted_month")
```

```python
top_5 = (df_us_merged
          .sum()
          .sort_values(ascending=False)
          .head(5)
          .index
          .to_list())
top_5
```

```python
df_us_merged[top_5].plot(kind='line')
plt.title('Montly Job Postings For Tech Jobs In US')
plt.xlabel('Year 2023')
plt.ylabel('Job Count')
plt.ylim(0,20000)
plt.legend()
plt.show()
```
